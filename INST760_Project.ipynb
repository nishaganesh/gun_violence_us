```python
#convert number of guns involved to numeric - it shouldn't be float
gun_violence['n_guns_involved'] = gun_violence['n_guns_involved'].fillna(-1).astype(int)

#convert congressional_district, state_house_district, state_senate_district to objects
gun_violence[['congressional_district', 'state_house_district', 'state_senate_district']] = gun_violence[['congressional_district', 'state_house_district', 'state_senate_district']].astype(object)
gun_violence.info()
```

```python
gun_violence.head()
```

```python
# creating month, year, weekday, week and quarter column
gun_violence['date'] = pd.to_datetime(gun_violence['date'])
gun_violence['year'] = gun_violence['date'].dt.year
gun_violence['month'] = gun_violence['date'].dt.month
gun_violence['weekday'] = gun_violence['date'].dt.weekday
gun_violence['week'] = gun_violence['date'].dt.week
gun_violence['quarter'] = gun_violence['date'].dt.quarter
gun_violence['killed_injured'] = gun_violence['n_killed'] + gun_violence['n_injured']
```

```python
gun_violence.head()
```

```python
gun_violence["participant_gender"] = gun_violence["participant_gender"].fillna("0::Unknown")


def clean_participant_gender(row) :
    gender_row_values = []
    gender_row = str(row).split("||")
    for x in gender_row :
        gender_row_value = str(x).split("::")
        if len(gender_row_value) > 1 :
            gender_row_values.append(gender_row_value[1])
            
    return gender_row_values


participant_genders = gun_violence.participant_gender.apply(clean_participant_gender)
gun_violence["participant_gender_total"] = participant_genders.apply(lambda x: len(x))
gun_violence["participant_gender_male"] = participant_genders.apply(lambda x: x.count("Male"))
gun_violence["participant_gender_female"] = participant_genders.apply(lambda x: x.count("Female"))
gun_violence["participant_gender_unknown"] = participant_genders.apply(lambda x: x.count("Unknown"))
del(participant_genders)
```

```python
gun_violence.head()
```

```python
gun_violence["n_guns_involved"] = gun_violence["n_guns_involved"].fillna(0)
gun_violence["gun_stolen"] = gun_violence["gun_stolen"].fillna("0::Unknown")
# Prints a lot but gives all the unique values of a column
#dataset_gunviolence["gun_stolen"].unique()

def clean_gun_stolen(row) :
    unknownCount = 0
    stolenCount = 0
    notstolenCount = 0
    gunstolen_row_values = []
    
    gunstolen_row = str(row).split("||")
    for x in gunstolen_row :
            gunstolen_row_value = str(x).split("::")
            if len(gunstolen_row_value) > 1 :
                gunstolen_row_values.append(gunstolen_row_value[1])
                if "Stolen" in gunstolen_row_value :
                    stolenCount += 1
                elif "Not-stolen" in gunstolen_row_value :
                    notstolenCount += 1
                else :
                    unknownCount += 1
                    
    return gunstolen_row_values


gunstolenvalues = gun_violence.gun_stolen.apply(clean_gun_stolen)
gun_violence["gun_stolen_stolen"] = gunstolenvalues.apply(lambda x: x.count("Stolen"))
gun_violence["gun_stolen_notstolen"] = gunstolenvalues.apply(lambda x: x.count("Not-stolen"))
del(gunstolenvalues)
```

```python
gun_violence.head()
```

```python
#gun_violence.participant_age_group.unique()
gun_violence["participant_age_group"] = gun_violence["participant_age_group"].fillna("0::Unknown")

def clean_participant_age_group(row) :
    unknownCount = 0
    childCount = 0
    teenCount = 0
    adultCount = 0
    agegroup_row_values = []
    
    agegroup_row = str(row).split("||")
    for x in agegroup_row :
        agegroup_row_value = str(x).split("::")
        if len(agegroup_row_value) > 1 :
            agegroup_row_values.append(agegroup_row_value[1])
            if "Child 0-11" in agegroup_row_value :
                childCount += 1
            elif "Teen 12-17" in agegroup_row_value :
                teenCount += 1
            elif "Adult 18+" in agegroup_row_value :
                adultCount += 1
            else :
                unknownCount += 1
                
    return agegroup_row_values

agegroupvalues = gun_violence.participant_age_group.apply(clean_participant_age_group)
gun_violence["agegroup_child"] = agegroupvalues.apply(lambda x: x.count("Child 0-11"))
gun_violence["agegroup_teen"] = agegroupvalues.apply(lambda x: x.count("Teen 12-17"))
gun_violence["agegroup_adult"] = agegroupvalues.apply(lambda x: x.count("Adult 18+"))
del(agegroupvalues)
```
